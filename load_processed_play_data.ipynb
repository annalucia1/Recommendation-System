```python
import pandas as pd

schema = ['uid','device','song_id','song_type','song_name','singer','play_time','song_length','paid_flag','file_name']
df = pd.read_csv('../data/play/20170401_3_play.log.fn',delimiter='\t',header=None,index_col=None,names=schema)
```

```python
df
```

```python
def get_date(file_name):
    tmp_list = str(file_name).split('_')
    return tmp_list[0]
df['date'] = df['file_name'].map(get_date)
```

```python
res = df[['uid','song_name']].groupby('song_name').count().rename(columns={'uid': 'count'})
res.sort_values(['count'], ascending=[0])
```

```python
res = df[['uid','date','song_id']].groupby(['uid','date']).count().rename(columns={'song_id': 'count'})
print res
```

```python
import math
def get_ceil_bin(val):
    return int(math.ceil(val/10)*10)
res['count_bin'] = res['count'].map(get_ceil_bin)
print res
```

```python
number_of_songs_listened_dist = res.groupby('count_bin').count()
print number_of_songs_listened_dist
```

```python
import matplotlib.pyplot as plt
plt.figure()
number_of_songs_listened_dist['count_bin']=number_of_songs_listened_dist.index
x = number_of_songs_listened_dist[:16]['count_bin'].values
y = number_of_songs_listened_dist[:16]['count'].values
plt.bar(x,y,width=8,color='g')
plt.show()
```
